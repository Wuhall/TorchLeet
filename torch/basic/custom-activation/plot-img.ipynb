```python
import numpy as np
import matplotlib.pyplot as plt

def plot_activation(func, x_range=(-5, 5), num_points=500, title=None):
    """
    绘制任意激活函数图像
    
    参数:
        func: 要绘制的函数
        x_range: x轴范围，默认为(-5, 5)
        num_points: 采样点数，默认为500
        title: 图像标题，默认为函数表达式
    """
    x = np.linspace(x_range[0], x_range[1], num_points)
    y = func(x)
    
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, linewidth=2)
    
    if title is None:
        title = f'Activation Function: {func.__name__}'
    plt.title(title, fontsize=16)
    plt.xlabel('x', fontsize=14)
    plt.ylabel('activation(x)', fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

# 定义你的激活函数
def tanh_plus_x(x):
    return np.tanh(x) + x

# 使用函数绘制图像
plot_activation(tanh_plus_x)
```
